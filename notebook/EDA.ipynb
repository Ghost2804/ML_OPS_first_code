```python
df = df.rename(columns={'DEATH_EVENT':'death_event'})
```

```python
num_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
cat_cols = df.select_dtypes(include='int64').columns.difference(['time', 'DEATH_EVENT'])

for col in cat_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col)
    plt.title(f'Count of {col}')
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
for col in df.columns.difference(['death_event']):
    plt.figure(figsize=(6, 4))
    if df[col].nunique() < 10:
        sns.countplot(data=df, x=col, hue='death_event')
    else:
        sns.boxplot(data=df, x='death_event', y=col)
    plt.title(f'{col} vs death_event')
    plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='time', hue='death_event', bins=30, kde=True)
plt.title('Time vs death_event')
plt.show()
```

```python
from scipy.stats import ttest_ind

dead = df[df['death_event'] == 1]['ejection_fraction']
alive = df[df['death_event'] == 0]['ejection_fraction']
t_stat, p_val = ttest_ind(dead, alive)

print(f"T-test: t={t_stat:.2f}, p={p_val:.4f}")
```
